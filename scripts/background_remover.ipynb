```python
# Importing necessary libraries
from PIL import Image
import numpy as np
from skimage.filters import threshold_otsu
import matplotlib.pyplot as plt

# Function for background subtraction
def background_subtraction(image_path):
    img = Image.open(image_path)
    grayscale_img = img.convert("L")  # Convert image to grayscale
    img_array = np.array(grayscale_img)  # Convert image to numpy array
    
    # Apply Otsu's threshold
    thresh_value = threshold_otsu(img_array)
    binary_image = img_array > thresh_value  # Generate binary image based on threshold
    
    return binary_image

image_path = "../assets/test.jpg"  # Update with the correct path to your image
binary_image = background_subtraction(image_path)

# Displaying results
plt.imshow(binary_image, cmap='gray')  # Display the binary image
plt.title("Background Subtracted Image")  # Title for the plot
plt.axis('off')  # Hide axes for better visualization
plt.show()
```
